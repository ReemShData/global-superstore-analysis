# Global Superstore Sales Analysis

## Project Overview
This project analyzes global sales data using Python, SQL, and Excel.

The analysis includes:
- Data Cleaning
- Exploratory Data Analysis (EDA)
- Data Visualization
- Business Insights

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
path = "/kaggle/input/datasets/reemalmutairi1/superstore/superstore.csv"

df = pd.read_csv(path)

df.head()
```

# Data Cleaning

```python
df.drop('记录数', axis=1, inplace=True)
```

```python
df['Order.Date'] = pd.to_datetime(df['Order.Date'])

df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])
```

# Dataset Information

```python
df.info()
```

# Top Countries by Sales

```python
top_countries = df.groupby('Country')['Sales'].sum().sort_values(ascending=False).head(10)

top_countries.plot(kind='bar', figsize=(8,5))

plt.title('Top 10 Countries by Sales')

plt.xlabel('Country')

plt.ylabel('Sales')

plt.xticks(rotation=45)

plt.show()
```

# Profit by Category

```python
profit_category = df.groupby('Category')['Profit'].sum()

profit_category.plot(kind='bar', figsize=(7,5))

plt.title('Profit by Category')

plt.xlabel('Category')

plt.ylabel('Profit')

plt.xticks(rotation=0)

plt.show()
```

This chart shows sales performance across different months.

# Monthly Sales Trend

```python
monthly_sales = df.groupby(df['Order.Date'].dt.month)['Sales'].sum()

monthly_sales.plot(kind='line', marker='o', figsize=(8,5))

plt.title('Monthly Sales Trend')

plt.xlabel('Month')

plt.ylabel('Sales')

plt.show()
```

This analysis explores the relationship between discount levels and profitability.

# Discount vs Profit

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Discount', y='Profit', data=df)

plt.title('Discount vs Profit')

plt.show()
```

# Shipping Cost vs Profit

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Shipping.Cost', y='Profit', data=df)

plt.title('Shipping Cost vs Profit')

plt.show()
```

# Top 10 Products by Sales

```python
top_products = df.groupby('Product.Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='barh', figsize=(10,6))

plt.title('Top 10 Products by Sales')

plt.xlabel('Sales')

plt.ylabel('Product')

plt.show()
```

# Key Insights

## Key Findings

- Technology category generated the highest profit.
- The United States achieved the highest sales.
- High discounts often reduce profit.
- Some products generate significantly higher sales than others.
- Shipping cost has a relationship with profitability.
